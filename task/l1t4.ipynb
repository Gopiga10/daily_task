```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
data = pd.DataFrame({
    "ID": np.arange(1, 101),
    "Category": np.random.choice(["A", "B", "C"], size=100),
    "Income": np.random.randint(20000, 80000, size=100),
    "Age": np.random.randint(20, 60, size=100)  
})
stratified_sample = data.groupby("Category", group_keys=False).apply(
    lambda x: x.sample(n=7, random_state=42), include_groups=False
)
fig, ax = plt.subplots(1, 3, figsize=(18, 6))
random_sample = data.sample(n=20, random_state=42)
sns.kdeplot(data["Income"], ax=ax[0], fill=True, color="purple", label="Original Data")
sns.kdeplot(random_sample["Income"], ax=ax[0], fill=True, color="orange", label="Random Sample")
ax[0].set_title("Random Sampling", fontsize=12, fontweight="bold")
ax[0].legend()
sns.kdeplot(data["Income"], ax=ax[1], fill=True, color="gray", label="Original Data")
sns.kdeplot(stratified_sample["Income"], ax=ax[1], fill=True, color="blue", label="Stratified Sample")
ax[1].set_title("Stratified Sampling", fontsize=12, fontweight="bold")
ax[1].legend()
systematic_sample = data.iloc[::5]  # Selecting every 5th row as a systematic sample
sns.kdeplot(data["Income"], ax=ax[2], fill=True, color="green", label="Original Data")
sns.kdeplot(systematic_sample["Income"], ax=ax[2], fill=True, color="red", label="Systematic Sample")
ax[2].set_title("Systematic Sampling", fontsize=12, fontweight="bold")
ax[2].legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
plt.hist(data["Age"], bins=10, alpha=0.5, color="gray", label="Original Data")
plt.hist(stratified_sample["Age"], bins=10, alpha=0.7, color="blue", label="Stratified Sample")
plt.xlabel("Age", fontsize=11)
plt.ylabel("Frequency", fontsize=11)
plt.title(" Age Distribution - Original vs Stratified Sample", fontsize=12, fontweight="bold", color="darkblue")
plt.legend()
plt.grid(alpha=0.3)
plt.show()


print("\n Cluster Sample Data (First 10 Rows):")
display(cluster_sample[["ID", "Age", "Income", "Category", "Sampling_Method"]])
```
