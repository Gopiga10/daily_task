```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_tested.csv") 

df["Sex"] = df["Sex"].map({"male": 0, "female": 1})

df = df.fillna({"Age": df["Age"].median(), "Fare": df["Fare"].median()})

numerical_df = df.drop(columns=["Name", "Ticket", "Embarked", "Cabin"], errors="ignore")
```

```python
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap="YlGnBu", 
    fmt=".2f", 
    linewidths=1.5, 
    cbar_kws={"shrink": 0.8, "label": "Correlation Strength"},
    vmin=-1, vmax=1
)
plt.title("Correlation Matrix of Titanic Dataset", fontsize=14, pad=15, weight="bold")
plt.show()
```
