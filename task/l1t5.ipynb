```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
file_path = "Mall_Customers.csv"
df = pd.read_csv(file_path)
print(df.head())
numerical_cols = ["Age", "Annual Income (k$)", "Spending Score (1-100)"]
cov_matrix = df[numerical_cols].cov()
corr_matrix = df[numerical_cols].corr(method="pearson")
print("\nCovariance Matrix:")
display(cov_matrix.style.set_caption("Covariance Matrix").format("{:.2f}").background_gradient(cmap="Blues"))
print("\nCorrelation Matrix:")
display(corr_matrix.style.set_caption("Correlation Matrix").format("{:.2f}").background_gradient(cmap="coolwarm"))
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Pearson's Correlation Coefficient", fontsize=12, fontweight="bold", color="darkblue")
plt.show()
```
