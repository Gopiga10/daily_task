```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
fig, axes = plt.subplots(2, 3, figsize=(15, 10), facecolor="#F5F5F5")
fig.suptitle("Visualization of Different Probability Distributions", fontsize=16, fontweight="bold", color="darkblue")
x = np.linspace(0, 10, 1000)
uniform_pdf = stats.uniform.pdf(x, loc=2, scale=6)  
axes[0, 0].plot(x, uniform_pdf, color="blue", linewidth=2)
axes[0, 0].fill_between(x, uniform_pdf, color="blue", alpha=0.3)
axes[0, 0].set_title("Uniform Distribution", fontsize=12, fontweight="bold")

# Normal Distribution
x = np.linspace(-4, 4, 1000)
normal_pdf = stats.norm.pdf(x, loc=0, scale=1)
axes[0, 1].plot(x, normal_pdf, color="red", linewidth=2)
axes[0, 1].fill_between(x, normal_pdf, color="red", alpha=0.3)
axes[0, 1].set_title("Normal Distribution", fontsize=12, fontweight="bold")

# Binomial Distribution
n, p = 10, 0.5  
x = np.arange(0, n + 1)
binomial_pmf = stats.binom.pmf(x, n, p)
axes[0, 2].bar(x, binomial_pmf, color="green", edgecolor="black", alpha=0.7)
axes[0, 2].set_title("Binomial Distribution", fontsize=12, fontweight="bold")

# Poisson Distribution
lambda_ = 3  
x = np.arange(0, 15)
poisson_pmf = stats.poisson.pmf(x, lambda_)
axes[1, 0].bar(x, poisson_pmf, color="purple", edgecolor="black", alpha=0.7)
axes[1, 0].set_title("Poisson Distribution", fontsize=12, fontweight="bold")

# Log-normal Distribution
x = np.linspace(0.1, 5, 1000)
lognormal_pdf = stats.lognorm.pdf(x, s=0.5, scale=np.exp(1))
axes[1, 1].plot(x, lognormal_pdf, color="orange", linewidth=2)
axes[1, 1].fill_between(x, lognormal_pdf, color="orange", alpha=0.3)
axes[1, 1].set_title("Log-normal Distribution", fontsize=12, fontweight="bold")
axes[1, 2].axis("off")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```
