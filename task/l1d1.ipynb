```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
n = 10    
p = 0.5  
x = np.arange(0, n + 1)
pmf_values = stats.binom.pmf(x, n, p)
plt.figure(figsize=(10, 6), facecolor="#F5F5F5")  
colors = plt.cm.viridis(np.linspace(0.2, 1, len(x)))  
plt.bar(x, pmf_values, color=colors, edgecolor="black", alpha=0.85)
for i in range(len(x)):
    plt.text(x[i], pmf_values[i] + 0.005, f'{pmf_values[i]:.2f}', 
             ha='center', fontsize=10, fontweight='bold', color="black")
plt.grid(axis='y', linestyle='--', alpha=0.6) 
plt.xlabel("Number of Successes", fontsize=12, fontweight='bold', color="darkblue")
plt.ylabel("Probability", fontsize=12, fontweight='bold', color="darkred")
plt.title("Binomial Distribution PMF (n=10, p=0.5)", fontsize=14, fontweight='bold', color="purple")
plt.xticks(x, fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0   
sigma = 1
x = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(x, mu, sigma)
plt.figure(figsize=(10, 6), facecolor="#F5F5F5") 
plt.plot(x, pdf_values, color="darkblue", linewidth=2.5, label="Normal PDF (μ=0, σ=1)")
plt.fill_between(x, pdf_values, color="skyblue", alpha=0.4)
plt.axvline(mu, color='red', linestyle='--', linewidth=2, label="Mean (μ = 0)")
plt.text(mu + 0.2, 0.35, "Mean (μ)", fontsize=12, fontweight='bold', color='red')
plt.xlabel("X - Values", fontsize=12, fontweight='bold', color="darkblue")
plt.ylabel("Density", fontsize=12, fontweight='bold', color="darkred")
plt.title(" Probability Density Function (PDF) of Normal Distribution", fontsize=14, fontweight='bold', color="purple")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.legend(loc="upper left", fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0   
sigma = 1 
x = np.linspace(-4, 4, 1000)
cdf_values = stats.norm.cdf(x, mu, sigma)
plt.figure(figsize=(10, 6), facecolor="#F5F5F5") 
plt.plot(x, cdf_values, color="darkgreen", linewidth=2.5, label="Normal CDF (μ=0, σ=1)")
plt.fill_between(x, cdf_values, color="lightgreen", alpha=0.4)
for point in [-2, 0, 2]: 
    plt.scatter(point, stats.norm.cdf(point, mu, sigma), color="red", s=100, edgecolors="black", zorder=3)
    plt.text(point, stats.norm.cdf(point, mu, sigma) - 0.05, f"P(X ≤ {point}) = {stats.norm.cdf(point, mu, sigma):.2f}",
             fontsize=12, fontweight='bold', color='black', ha='center')
plt.xlabel("X - Values", fontsize=12, fontweight='bold', color="darkblue")
plt.ylabel("Cumulative Probability", fontsize=12, fontweight='bold', color="darkred")
plt.title(" Cumulative Distribution Function (CDF) of Normal Distribution", fontsize=14, fontweight='bold', color="purple")
plt.grid(axis='y', linestyle='--', alpha=0.6)  
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.legend(loc="lower right", fontsize=12)
plt.show()
```
